convert age to number

```python
Df[' AGE'] = pd.to_numeric(Df[' AGE'], errors='coerce')
```

fill missing age with median

```python
Df[' AGE'] = Df[' AGE'].fillna(Df[' AGE'].median())
```

```python
Df.head(10)
```

Lower case and remove space

```python
Df.columns = Df.columns.str.strip().str.lower().str.replace(' ', '_')
```

```python
Df['country'] = Df['country'].str.strip().str.lower().replace({
    'usa': 'united states',
    'u.s.': 'united states',
    'united states': 'united states',
    'uk': 'united kingdom'
}).fillna('unknown')
```

```python
Df.head(15)
```

for gender column it wil convert m to male and f to female

```python
Df['gender'] = Df['gender'].str.strip().str.lower().replace({
    'm': 'male',
    'f': 'female',
    '': 'unknown',
    None: 'unknown'
}).fillna('unknown')
```

```python
Df.head(10)
```

for email column

```python
Df['email'] = Df['email'].str.strip().str.lower().fillna('unknown')
```
